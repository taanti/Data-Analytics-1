## Data Imputation

Data imputation refers to the process of filling in missing values in a dataset. This is a common task in data preprocessing, especially when dealing with real-world datasets that often have missing or incomplete information. 

```python
#Show the column visualization and check the data distribution

df.Age.plot(kind='hist');
```

Because column is **skewness** distributed then we will do imputation on column by using median

```python
val = df.Age.median()
df['Age'] = df.Age.fillna(val)
```

show dataset info to see if the column has been imputed, it turns out that the column has now changed its number

```python
df.info()
```

## Drop Data
Dropping data involves removing rows or columns from a dataset, typically due to reasons like missing values, outliers, redundant variables, data quality issues, or specific sample selection criteria. It should be done carefully, as it may lead to information loss or biased results.

```python
#Load csv data

df = pd.read_csv('Dataset/Lusi.csv')
df
```

**Example** : the total number of data entries is 1961, while the Status column is 791 it means there is null data in the Status column

```python
#show the proportion of Status column data
df.Position.value_counts()
```

```python
##Check how many of missing values in column
df['Position'].isna().sum()
```

```python
#calculates the percentage of missing values in column
df['Position'].isna().sum()/df.shape[0] * 100
```

It can be seen that the Status value column has too much missing data and also the info column Status is not too informative to know the data, then we will delete the Status column

```python
df.drop('Position', axis=1, inplace = True)
```

show dataset info to see if the column has been deleted, it turns out that the column is no longer displayed

```python
df.info()
```

## Categorical Nominal

Categorical nominal data refers to variables in a dataset that represent categories or groups with no inherent order or ranking. These categories are distinct and separate, and there is no natural ordering among them.

```python
df['Citizenship'].isna().sum()
```

```python
df['Citizenship'].value_counts()
```

By checking the data type of the “Citizenship” column first and knowing that it is categorical data, the next step is to use the mode to fill in the missing values.

```python
val = df.Citizenship.mode().values[0]
df['Citizenship'] = df.Citizenship.fillna(val)
```

after imputation, it can be seen that the proportion has changed.

```python
df['Citizenship'].value_counts()
```

### Pivot

Pivoting is a data manipulation technique used to reshape data from a long format to a wide format or vice versa. This process involves rotating the data to reorganize it according to different dimensions or variables.

```python
#Pivot Table Visualization
def make_pivot(param1, param2):
    df_slice = df[[param1, param2, 'Passenger ID']]
    slice_pivot = df_slice.pivot_table(index=[param1], columns =[param2], \
                                      aggfunc=np.size,fill_value=0)
    p_chart= slice_pivot.plot(kind='bar');
    for p in p_chart.patches:
        p_chart.annotate(str(p.get_height()), (p.get_x() * 1.05, p.get_height() * 1.01))
        return slice_pivot
        return p_chart

make_pivot('Sex', 'Fate')
```

#### Pivot Slice

Pivot slice, or sometimes referred to as pivot table slicing, is a technique used in data analysis to extract specific subsets of data from a pivot table based on certain criteria or conditions. It involves selecting and displaying only the relevant portions of the pivot table that meet the specified criteria.

```python
df_slice = df[['Fate', 'Passenger/Crew', 'Passenger ID']]
slice_pivot = df_slice.pivot_table(index=['Fate'], columns=['Passenger/Crew'], \
                                  aggfunc=np.size,fill_value=0)
slice_pivot
```

```python
df_slice = df[['Fate', 'Sex','Passenger/Crew', 'Passenger ID']]
slice_pivot = df_slice.pivot_table(index=['Fate', 'Sex'], columns=['Passenger/Crew'], \
                                  aggfunc=np.size,fill_value=0)
slice_pivot
```

It creates a contingency table that shows the frequency distribution of the data for each combination of categories from the variables involved.
